```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
customer_df.head()
```

```python
numericals = customer_df.select_dtypes(include=np.number)
numericals.head()
```

```python
categoricals = customer_df.select_dtypes(include=object)
categoricals.head()
```

```python
def plt_num(x):
    for i in x:
        sns.displot(x[i])
        plt.show()
plt_num(numericals)
```

we see from "months since policy inception" kind of uniform distrubition and maybe in "months since last claim" but others tend to show poisson distrubition.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
numericals_normalized = pd.DataFrame(MinMaxScaler().fit_transform(numericals), columns=numericals.columns)
numericals_normalized.shape
```

```python
plt_num(numericals_normalized)
```

```python
from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(numericals)
numericals_standardized = transformer.transform(numericals)
numericals_standardized = pd.DataFrame(numericals_standardized, columns=numericals.columns)
```

```python
plt_num(numericals_standardized)
```

```python
def log_tran(x):
    if  x!=0:
        return np.log(x+1)
    else:
        return np.log(x+1)
```

```python
for column in numericals:
    numericals[column]=numericals[column].apply(log_tran)
```

```python
numericals.head()
```

```python
plt_num(numericals)
```

In conclusion minmax scaler and standard scaler haven't changed the plotting. But logtransformation has changed the distrubution as we could expect.
